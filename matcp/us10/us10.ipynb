```python
import matplotlib.pyplot as plt
```

```python
# get equipment names and occurrences
equipmentFile = open("EquipmentUsed.csv", "r", encoding='utf-8')
lineCount = 0
txtLabels = {}
# skipping the first line as it is the header
next(equipmentFile)
for line in equipmentFile:
  equipmentPiece = line.strip() # sanitization
  lineCount += 1
  if equipmentPiece not in txtLabels.keys():
    txtLabels[equipmentPiece] = 1
  else:
    txtLabels[equipmentPiece] += 1
equipmentFile.close()
```

```python
plt.pie(list(txtLabels.values()), labels=list(txtLabels.keys()), autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
```
